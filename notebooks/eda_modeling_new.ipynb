```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample NRI dataset
data = {
    "monthly_income": [5000, 8000, 12000],
    "employment_years": [2, 4, 7],
    "remittance_amount": [1500, 2500, 4000],
    "credit_score": [1900, 3100, 4750]
}

df = pd.DataFrame(data)
df
```

```python
df["credit_score"].plot(kind="hist")
plt.title("NRI Credit Score Distribution")
plt.xlabel("Credit Score")
plt.show()
```
